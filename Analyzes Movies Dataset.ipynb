The goal of the project is to use python loops, Numpy functions, to manipulate on Pandas Dataframe derived from a csv file and find out more about movies' trend over the years. I also plotted a beautiful graph using Matplotlib. 

The columns of the dataset are as follows:

| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |
| `country` | Country of origin |
| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
netflix_df=pd.read_csv('netflix_data.csv')
netflix_subset=netflix_df[netflix_df['type']=='Movie']          ##Data Filtering to only movies in the dataset
netflix_movies=netflix_subset[['title','country','genre','release_year','duration']]   ##Data cleaning to keep only what I need for the project
short_movies=netflix_movies[netflix_movies['duration']<60]
print(short_movies)
colors=[]                                      ##Here, I created an empty list to store colors for different genres of the films to later plot in graph
for lab, row in netflix_movies.iterrows():     ##A for loop and if statements are used to assign colors based on different genres
    if row['genre']=='Stand-Up':
        col='blue'
    elif row['genre']=='Documentaries':
        col='yellow'
    elif row['genre']=='Children':
        col='red'
    else:
        col='purple'
    colors.append(col)
fig=plt.figure(figsize=(12,8))             ##plotting the graph using matplotlib
plt.scatter(netflix_movies['release_year'],netflix_movies['duration'],c=colors)
plt.xlabel('Release year')                 ##Labeling the graph
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')

##At the end of this project, I found out that documentaries, stand-up, and children's categories usually have shorter runtime than other genres 
##of movies. However, there is no clear conclusion on whether the films are getting shorter over time. The purple dots that represent other films have 
##a even longer run time, yet the netflix database shows that different variations of films have different outcomes in runtime.
```
